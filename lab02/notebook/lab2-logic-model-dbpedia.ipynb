## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
data_dbpedia = requests.get('http://dbpedia.org/data/Tent_of_Miracles_(novel).jsod').json()
print(json.dumps(data_dbpedia, indent=3))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres?

```python
author_url = data_dbpedia["d"]["results"][0]["http://dbpedia.org/ontology/author"]["__deferred"]["uri"]
print("Autor de Tenda dos Milagres: " + author_url.split('/')[-1].replace('_', ' '))
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
descriptor_aspirin = requests.get("https://id.nlm.nih.gov/mesh/lookup/descriptor?label=aspirin&match=exact&limit=10").json()
print(json.dumps(descriptor_aspirin, indent=3))
```

```python
code_aspirin = descriptor_aspirin[0]['resource'].split('/')[-1]
data_aspirin = requests.get(f'https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code_aspirin}').json()
print(json.dumps(data_aspirin['terms'], indent=3))
```

```python
print(f"Os termos para Aspirina são: {', '.join([term['label'] for term in data_aspirin['terms']])}")
```

## Tarefa 4
### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
data_osm = requests.get('http://nominatim.openstreetmap.org/search?format=json&q=Empire State Building').json()
print(json.dumps(data_osm, indent=3))
```

```python
way_esb = requests.get(f'https://api.openstreetmap.org/api/0.6/way/{data_osm[0]["osm_id"]}.json').json()
print(json.dumps(way_esb, indent=3))
```

```python
esb_wiki = way_esb["elements"][0]["tags"]["wikipedia"].replace(' ', '_').replace('en:', '')
esb_dbpedia = requests.get(f'http://dbpedia.org/data/{esb_wiki}.jsod').json()

print(json.dumps(esb_dbpedia, indent=3))
```
